```python
all_data = [ssim_data,psnr_data]
all_labels = ['SSIM','PSNR']
for data,data_lbl in zip(all_data,all_labels):
    labels = []
    input_avgs = []
    result_avgs = []
    filter_legends = []
    for sa in data.keys():
        #print(sa)
        sample = sa.split('_')[0]
        filter_lbl = sa.split('_')[1]+'%'
        input_np = np.load(data[sa]['input'])
        result_np = np.load(data[sa]['result'])
        labels.append(sample)
        filter_legends.append(filter_lbl)
        input_avgs.append(np.mean(input_np))
        result_avgs.append(np.mean(result_np))
    labels = np.array(labels)
    input_avgs = np.array(input_avgs)
    result_avgs = np.array(result_avgs)
    for i in range(3):
        print('generating '+data_lbl+' plot for sample '+str(i+1))
        plt.figure(figsize=(8.5, 6))
        plt.scatter(filter_legends[i*4:(i+1)*4],result_avgs[i*4:(i+1)*4], c='b', label='model result')
        plt.scatter(filter_legends[i*4:(i+1)*4],input_avgs[i*4:(i+1)*4], c='r',label='noisy input')
        axes = plt.gca()
        axes.set_xlim([-0.5,4])
        axes.legend(loc='center right')
        plt.title('MWCNN: '+data_lbl+' Values for each filter in sample '+str(i+1))
        for input_val,result_val,label in zip(input_avgs[i*4:(i+1)*4],result_avgs[i*4:(i+1)*4],filter_legends[i*4:(i+1)*4]):
            plt.annotate("%.4f" % input_val, (label, input_val+0.04))
            plt.annotate("%.4f" % result_val, (label, result_val-0.06))
        plt.savefig('MWCNN_'+data_lbl+'_Results_sample'+str(i+1)+'.png',dpi=600)
```
